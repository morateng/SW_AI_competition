```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
from sklearn.linear_model import LogisticRegression

drop_list = ['index', 'country', 'TIPI2', 'TIPI3', 'TIPI4', 'TIPI5', 'TIPI7', 'TIPI8', 'TIPI9', 'Q18', 'Q19']

train = train.drop(drop_list, axis=1)
test = test.drop(drop_list, axis=1)
```

```python
train_x = train.drop(['nerdiness'], axis=1)
train_y = train['nerdiness']
```

```python
train_x.head(5)
```

```python
train_x.info()
```

```python
train_x.isnull().sum()
```

```python
replace = train_x['Q1'].value_counts().index[0]
train_x['Q1'] = train_x['Q1'].fillna(replace)
test['Q1'] = test['Q1'].fillna(replace)

for i in train_x:
    replace = train_x.loc[:,i].value_counts().index[0]
    train_x[i] = train_x[i].fillna(replace)
    test[i] = test[i].fillna(replace)
```

```python
test.isnull().sum()
```

```python
train_x.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=0)
lr.fit(train_x, train_y)
```

```python
y_pred = lr.predict(test)
```

```python
submission = pd.read_csv('data/sample_submission.csv')
submission
```

```python
submission['nerdiness'] = y_pred
submission
```

```python
submission.to_csv('eda2.csv', index=False)
```
