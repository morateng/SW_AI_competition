```python
train0 = pd.read_csv('data/train.csv')
test0 = pd.read_csv('data/test.csv')

train0 = train0.drop(['index', 'country', 'TIPI2', 'TIPI3', 'TIPI4', 'TIPI5', 'TIPI7', 'TIPI8', 'TIPI9', 'Q18', 'Q19'], axis=1)
test0 = test0.drop(['index', 'country', 'TIPI2', 'TIPI3', 'TIPI4', 'TIPI5', 'TIPI7', 'TIPI8', 'TIPI9', 'Q18', 'Q19'], axis=1)
```

```python
train_x = train0.drop(['nerdiness'], axis=1)
train_y = train0['nerdiness']
```

```python
from lightgbm import LGBMClassifier

lgbm_clf = LGBMClassifier(n_estimators = 10000)
lgbm_clf.fit(train_x, train_y)

lgbm_pred = lgbm_clf.predict(test0)
```

```python
submission = pd.read_csv('data/sample_submission.csv')

submission
```

```python
lgbm_pred
```

```python
submission['nerdiness'] = lgbm_pred
```
